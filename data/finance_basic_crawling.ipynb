```python
import pandas as pd
# ticker 조회
# ticker 시작일부터 끝날 까지 조회후 kospi, kosdaq 확인

# get_market_ticker_list 메서드는 지정한 일자(YYYYMMDD)의 코스피 시장에 상장된 ticker를 리스트로 반환합니다.
# tickers = stock.get_market_ticker_list("20190225")
try:
    tickers_ksp = stock.get_market_ticker_list(endDay, market="KOSPI") # 코스닥 조회, 입력 안하면 KOSPI
    tickers_kdq = stock.get_market_ticker_list(endDay, market="KOSDAQ")
#     tickers_ksp=pd.DataFrame(tickers_ksp, columns= ["종목코드"])
#     tickers_kdq=pd.DataFrame(tickers_kdq, columns= ["종목코드"])
    print(tickers_ksp)
    print(tickers_kdq)
except Exception as e:
    print(e)
```

```python
# KSP 수정 종가 구하기
from pandas_datareader import data as pdr
import yfinance as yf
from tqdm import tqdm
yf.pdr_override()
dfs = []

for i in tqdm(range(len(tickers_ksp))):
    df = pdr.get_data_yahoo("{}.KS".format(tickers_ksp[i]), "2019-01-01", "2023-05-31")
    df['ticker'] = tickers_ksp[i]
    df_=df.drop('Open',axis = 1)
    df__ = df_.drop('High', axis = 1)
    df___ = df__.drop('Low', axis=1)
    df____ = df___.drop('Close', axis=1)
    df_____ = df____.drop('Volume', axis=1)
    dfs.append(df_____)
    
combined_df = pd.concat(dfs)
ksp_df=combined_df.reset_index()
print(ksp_df)

#     print(df)
```

```python
# Kdq 수정 종가 구하기

from pandas_datareader import data as pdr
import yfinance as yf
from tqdm import tqdm
yf.pdr_override()
dfs = []
if len(tickers_kdq) > 0:
    for i in tqdm(range(len(tickers_kdq))):
        df = pdr.get_data_yahoo("{}.KQ".format(tickers_kdq[i]), "2019-01-01", "2023-05-31")
        df['ticker'] = tickers_kdq[i]
        df_ = df.drop(['Open', 'High', 'Low', 'Close', 'Volume'], axis=1)
        dfs.append(df_)
    if len(dfs) > 0:
        combined_df = pd.concat(dfs)
        kdq_df = combined_df.reset_index()
        print(kdq_df)
    else:
        print("No data available for ticker symbols in tickers_kdq.")
else:
    print("The tickers_kdq list is empty.")
    

#     print(df)
```

```python
# KSP 시총 구하기

from pykrx import stock 
from tqdm import tqdm
import warnings
import pandas as pd

warnings.filterwarnings('ignore')

market_cap_ksp = []

for i in tqdm(range(len(tickers_ksp))):
    df = stock.get_market_cap("20190101", "20230531", tickers_ksp[i])
    df_ = df.drop(['거래량', '거래대금', '상장주식수'], axis=1)
    ksp_market = df_.rename(columns={'시가총액':'MarketCap'})
#     print(ksp_market)
    market_cap_ksp.append(ksp_market)
market_cap_ksp_ = pd.concat(market_cap_ksp)
print(market_cap_ksp_)

```

```python
# KDQ 시총 구하기

from pykrx import stock 
from tqdm import tqdm
import warnings
import pandas as pd

warnings.filterwarnings('ignore')

market_cap_kdq = []

for i in tqdm(range(len(tickers_kdq))):
    df = stock.get_market_cap("20190101", "20230531", tickers_kdq[i])
    df_ = df.drop(['거래량', '거래대금', '상장주식수'], axis=1)
    kdq_market = df_.rename(columns={'시가총액':'MarketCap'})
#     print(kdq_market)
    market_cap_kdq.append(kdq_market)
    
market_cap_kdq_ = pd.concat(market_cap_kdq)
print(market_cap_kdq_)

```
