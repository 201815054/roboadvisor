```python
train_ran = int(len(df)*0.6)
valid_ran = int(len(df)*0.2)
print(train_ran)
print(valid_ran)
```

```python
train = df.iloc[:train_ran,:]
valid = df.iloc[train_ran:train_ran+valid_ran,:]
test = df.iloc[train_ran+valid_ran:,:]
train
```

```python
# 스케일링
def min_max_normal(tmp_df):
    eng_list = []
    sample_df = tmp_df.copy()
    for x in sample_df.columns:
        if x == 'date' or x == 'name' or x == 'code':
            continue
        series = sample_df[x].copy()
        values = series.values
        values = values.reshape((len(values),1))
        # 스케일러 생성 및 훈련
        scaler = MinMaxScaler(feature_range=(0,1))
        scaler = scaler.fit(values)
        # 데이터셋 정규화 및 출력 
        normalized = scaler.transform(values)
        new_feature = f'{x}_normal'
        eng_list.append(new_feature)
        sample_df[new_feature] = normalized
    return sample_df, eng_list 
```

```python
train_sample_df, eng_list = min_max_normal(train)
val_sample_df, eng_list = min_max_normal(valid)
test_sample_df, eng_list = min_max_normal(test)
```

```python
def create_dataset_binary(data, feature_list, step, n):
    # train_xdata = np.array(data[feature_list[0:n]]) # 종가떨어뜨려야함
    train_xdata = np.array(data.drop('close_normal',axis=1)) # 종가떨어뜨려야함

    m = np.arange(len(train_xdata) - step)
    x, y = [],[]
    for i in m:
        a = train_xdata[i:(i+step)]
        x.append(a)
    x_batch = np.reshape(np.array(x), (len(m),step,n)) # 종가를뺀 컬럼갯수 n 
    train_ydata = np.array(data[[feature_list[n]]])
    for i in m + step:
        start_price = train_ydata[i-1][0]
        end_price = train_ydata[i][0]

        if end_price > start_price:
            label = 1
        else:
            label = 0
        y.append(label)
    
    y_batch = np.reshape(np.array(y), (-1,1))
    return x_batch, y_batch
```

```python
num_step = 5
num_unit = 200 
n_feature = len(eng_list)-1

x_train, y_train = create_dataset_binary(train_sample_df[eng_list], eng_list, num_step, n_feature)
x_val, y_val = create_dataset_binary(val_sample_df[eng_list], eng_list, num_step, n_feature)
x_test, y_test = create_dataset_binary(test_sample_df[eng_list],eng_list, num_step, n_feature)
```

```python
x_train.shape
```

```python
# 모델 구조 
from tensorflow.keras.models import Model,load_model
from tensorflow.keras.layers import Input, Dense, LSTM
from tensorflow.keras.layers import Activation, BatchNormalization
from tensorflow.keras.layers import Dropout
from tensorflow.keras import backend as K
from tensorflow.keras import regularizers
```

```python
# LSTM 모델 생성 
K.clear_session()
input_layer = Input(batch_shape=(None,x_train.shape[1],x_train.shape[2]))
layers_lstm_1 = LSTM(num_unit, return_sequences = True, recurrent_regularizer=regularizers.L2(0.01))(input_layer)
layers_lstm_1 = BatchNormalization()(layers_lstm_1)
layers_lstm_2 = LSTM(num_unit, return_sequences= True, recurrent_regularizer=regularizers.L2(0.01))(layers_lstm_1)
layers_lstm_2 = Dropout(0.25)(layers_lstm_2)
layers_lstm_3 = LSTM(num_unit, return_sequences= True, recurrent_regularizer=regularizers.L2(0.01))(layers_lstm_2)
layers_lstm_3 = BatchNormalization()(layers_lstm_3)
layers_lstm_4 = LSTM(num_unit, return_sequences = True, recurrent_regularizer=regularizers.L2(0.01))(layers_lstm_3)
layers_lstm_4 = Dropout(0.25)(layers_lstm_4)
layers_lstm_5 = LSTM(num_unit, recurrent_regularizer= regularizers.L2(0.01))(layers_lstm_4)
layers_lstm_5 = BatchNormalization()(layers_lstm_5)
output_layer = Dense(1, activation = 'sigmoid')(layers_lstm_5)

model = Model(input_layer, output_layer)
model.compile(loss = 'binary_crossentropy', optimizer ='adam', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train,
                    y_train,
                    epochs = 20,
                    batch_size = 10,
                    verbose = 1,
                    validation_data = (x_val, y_val))
```

```python
y_val.shape
```

```python
# 결과값 그래프그리기
def plot_history(history):
    plt.figure(figsize=(15,5))
    ax = plt.subplot(1,2,1)
    plt.plot(history.history["loss"])
    plt.title("Train loss")
    ax = plt.subplot(1,2,2)
    plt.plot(history.history["val_loss"])
    plt.title("Test loss")
plot_history(history)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
# 예측
predicted = model.predict(x_test)
y_pred = [1 if i > 0.5 else 0 for i in predicted]
Y_test = [1 if i > 0.5 else 0 for i in y_test]
cm = confusion_matrix(Y_test,y_pred)
report = classification_report(Y_test,y_pred)
```
